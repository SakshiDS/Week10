### 1. Write a function that calculates the dot product manually AND a function that calculates the cross product manually. Both should accept two vectors and perform that calculation. Be sure to use try except and error handling for basic issues that might occur.


```python
# Defining function for dot product
def dotprod(a,b):
    try:
        if len(a)<len(b):
            r=len(a)
        else:
            r=len(b)

            dot = 0
            for i in range(r):
                dot = dot + a[i]*b[i]
            print(dot)
    except Exception as e:
        print(e)
```

```python
a=np.array([2,4,6])
b=np.array([5,2,-1])

dotprod(a,b)
```

```python
# Defining function for cross product
def crossprod(a,b):  
    dimension = len(a)
    if 1 < dimension < 4:
        cross = []
        try:
            for i in range(dimension):
                if i == 0:
                    j,k = 1,2
                    cross.append(a[j]*b[k] - a[k]*b[j])
                elif i == 1:
                    j,k = 2,0
                    cross.append(a[j]*b[k] - a[k]*b[j])
                else:
                    j,k = 0,1
                    cross.append(a[j]*b[k] - a[k]*b[j])
            return cross
        except Exception as e:
            print(e)
    else:
        print("incompatible dimensions for cross product (dimension must be 2 or 3)")
```

```python
a=np.array([2,4,6])
b=np.array([5,2,-1])

crossprod(a,b)
```

### 2. In a markdown section in your notebook, explain what one hot encoding is and why we use it. Then, manually calculate a one hot encoding for the Belize_Airbnb_listings.csv neighborhood_group column. Note: manually means donâ€™t use scikitlearn or keras.

```python
belize_df = pd.read_csv("belize_airbnb_listings.csv")
```

```python
# Identifying the unique neighbourhood groups and saving it as seperate list
data = belize_df["neighbourhood_group"].unique()
```

```python
print(data)
```

```python
# Mapping integer to each unique neighborhood group and creating a reverse map as well
group_to_code = dict((group, i) for i, group in enumerate(data))
code_to_group = dict((i, group) for i, group in enumerate(data))
```

```python
# encoding the column values
group_encoded = [group_to_code[group] for group in belize_df["neighbourhood_group"]]
print(group_encoded)
```

```python
# one hot encode
onehot_encoded = list()
for value in group_encoded:
    #Creating matrix of size 6 of zeroes
    neighborhood = [0 for _ in range(len(data))]
    #Inserting 1 for the code value
    neighborhood[value] = 1
    #Appending to the list of one hote encode
    onehot_encoded.append(neighborhood)
print(onehot_encoded)
```

```python
# invert encoding
inverted=list()
for i in range(len(onehot_encoded)):
    invert = code_to_group[np.argmax(onehot_encoded[i])]
    inverted.append(invert)
print(inverted)
```

```python
# Reverting from code and checking if its equal to the orginal column values. 
inverted_df = pd.DataFrame(inverted)
belize_df["neighbourhood_group"].equals(inverted_df[0])
```
